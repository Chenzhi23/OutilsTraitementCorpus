Dans le cas de nos données, nous les avons bien structurées au format **csv**

Python a une bibliothèque expres pour lire les csv

```python
import csv

with open("../../data/data.csv", 'r') as file:
    data = csv.reader(file, delimiter=";")
    for row in data:
        print(row)
```
